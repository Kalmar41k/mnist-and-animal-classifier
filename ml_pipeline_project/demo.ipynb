## In this demo, we will check how the two models cooperate with each other.

```python
from pipeline import AnimalVerificationPipeline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

Creating a model pipeline

```python
pipeline = AnimalVerificationPipeline()
```

Creating a text to predict

```python
text = "The image shows a cow."
```

Choosing an image to predict

```python
image_path = "data/Cow.jpg"
```

Visualizing the image

```python
def show_image(img_path):
    img = mpimg.imread(img_path)
    plt.imshow(img)
    plt.axis("off")
    plt.show()
```

```python
show_image(image_path)
```

```python
result = pipeline.verify_animal(text, image_path)
```

```python
print(result)
```

### The result showed that the text "The image shows a cow." contains an animal that is depicted in the image we selected. This means that our pipeline is working successfully

Since our image dataset contains only ten animals (butterfly, cat, chicken, cow, dog, elephant, horse, sheep, spider, squirrel), we will not be able to get a correct prediction if we select an image of another animal that is not in this list.

```python
text = "I see a fish in an aquarium."
image_path = "data/fish.jpg"
show_image(image_path)
```

```python
result = pipeline.verify_animal(text, image_path)
print(result)
```
